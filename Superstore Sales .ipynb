```python
# Convert date columns to datetime 
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Order Date'])
```

## 6. Exploratory Data Analysis (EDA)

### a. Sales by Category

```python
category_sales=df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales
```

```python
  #  Plot 
plt.figure(figsize=(8,4))
category_sales.plot(kind='bar',color='skyblue')
plt.title('Sales By Category')
plt.ylabel('Sales')
plt.savefig('Sales by Category.png',dpi=300,bbox_inches='tight')
plt.show()
```

### b. Top 10 Products by Sales

```python
top_products=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
top_products
```

```python
   # Plot 
top_products.plot(kind='bar',color='orange')
plt.title('Top 10 Products by Sales') 
plt.ylabel('Sales')
plt.savefig('Top 10 Products by Sales.png',dpi=300,bbox_inches='tight')
plt.show()
```

### c. Sales Trend Over Time

```python
df['Month']=df['Order Date'].dt.month
```

```python
monthly_sales=df.groupby('Month')['Sales'].sum()
```

```python
monthly_sales.plot(figsize=(10,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.savefig('Monthly Sales Trend.png',dpi=300,bbox_inches='tight')
plt.show()
```

### d. Profit by Region

```python
region_profit=df.groupby('Region')['Profit'].sum()
```

```python
sns.barplot(x=region_profit.index,y=region_profit.values)
plt.title('Profit by Region')
plt.ylabel('Profit')
plt.savefig('Profit by Region.png',dpi=300,bbox_inches='tight')
plt.show()
```

## 7. Advanced Analysis

### a. Correlation Analysis

```python
corr=df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
```

```python
sns.heatmap(corr,annot=True,cmap='Blues')
plt.title('Correlation Matrix')
plt.savefig('Corr.png',dpi=300, bbox_inches='tight')
plt.show()
```

### b. Customer Segmentation Example

```python
customers_sales=df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
     # Visualize
customers_sales.plot(kind='bar',color='purple')
plt.title('Top Customers by Sales')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.savefig('my_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
